```python
import numpy as np
import torch
import os

from scipy.spatial.distance import euclidean
from helpers.utils import load_pickle, save_pickle
import matplotlib.pyplot as plt

Online_result_save_rootdir = "./Online_experiments"
sub_name_online = "Jyt_test_0905_online_simu_1"
experiment_name = "lr0.001_dropout0.5"
result_save_subjectdir = os.path.join(Online_result_save_rootdir, sub_name_online, experiment_name)

memroyBank = load_pickle(result_save_subjectdir, "memoryBank_sources.pkl")

average_distances = []

for i,(key, value) in enumerate(memroyBank.items()):
    dist = []
    for j in range(value.shape[0]):
        for k in range(j+1, value.shape[0]):
            dist.append(np.linalg.norm(value[j].flatten() - value[k].flatten()))
    #print(len(dist))

    average_distances.append(np.mean(dist))


plt.plot(average_distances)
plt.show()
```
